```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data for a simple linear regression example
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualize the data
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data for Linear Regression")
plt.show()

# Model Representation
# We'll represent a linear regression model using the equation: y = mx + b
# where 'm' is the slope (coefficient) and 'b' is the intercept (bias).

# Initialize model parameters
m = 0
b = 0

# Define the hypothesis function
def hypothesis(X, m, b):
    return m * X + b

# Define the cost function (mean squared error)
def cost_function(X, y, m, b):
    m = len(y)
    predicted_values = hypothesis(X, m, b)
    cost = (1 / (2 * m)) * np.sum((predicted_values - y) ** 2)
    return cost

```
